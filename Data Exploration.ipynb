```python
#skewness of 1.88 and a kurtosis of 6.54 indicate that the distribution has positive skewness and high positive kurtosis, meaning it has a long right tail and a high peak.
print("Skewness: %f" % df_train['SalePrice'].skew())
print("Kurtosis: %f" % df_train['SalePrice'].kurt())
```

```python
#the skewness is 12.207688 which indicates a strong positive skewness and kurtosis is 203.243271 which indicates a very peaked distribution.
print("Skewness: %f" % df_train['LotArea'].skew())
print("Kurtosis: %f" % df_train['LotArea'].kurt())
```

```python
#When you have a strong skewness and peaked distribution in your data, it means that your data is not normally distributed and it can have an impact on your model performance. 
#Here are a few things you can do to handle this issue:

#Transformation: Apply a transformation such as logarithmic, square root or inverse transformation to the data to reduce skewness and kurtosis.

#Standardization: Standardizing the data using mean and standard deviation can also help in reducing skewness and kurtosis.

#Removing Outliers: Removing outliers from the data can also help in reducing skewness and kurtosis.

#Binning: Binning the data into smaller groups and converting it into a categorical variable can also help in reducing skewness and kurtosis.

#Regularization: Using regularization techniques like Lasso, Ridge or Elastic Net during model training can help in reducing the impact of skewed data on the model performance.
```

```python
# Correlation visualization
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
#The function np.corrcoef calculates the correlation matrix for the given data by taking the transpose of the data
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# Let's find out how missing data do we have

# First, let's count the number of null values
total = df_train.isnull().sum().sort_values(ascending=False)
# Then, let's calculate the percentage of missing data per feature
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
# Finally, let's concatenate Total and Percent into another dataframe
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
#standardizing data
#np.newaxis is a special object that can be used to increase the number of dimensions of an existing array. It is used to convert a 1D array into a 2D array, or to insert a new axis into an existing array.
saleprice_standardized = StandardScaler().fit_transform(df_train['SalePrice'][:,np.newaxis]);
#This code of line is taking a standardized SalePrice column from the training data and sorting it in ascending order using argsort(). 
#Then, it is slicing the first 20 values using the index [:20] and assigning it to the variable low_range. The purpose of this code is likely to extract the 20 lowest standardized SalePrice values for further analysis.
low_range = saleprice_standardized[saleprice_standardized[:,0].argsort()][:20]
#slicing the last 20 values using the index [:20] and assigning it to the variable high_range. 
high_range= saleprice_standardized[saleprice_standardized[:,0].argsort()][-20:]
print('outer range (low) of the distribution:')
low_range
```
